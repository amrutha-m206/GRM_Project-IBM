```python
import matplotlib.pyplot as plt
status_counts = df['Status'].value_counts()


plt.figure(figsize=(7,7))
status_counts.plot.pie( labels=[ 'Completed','In Progress'], colors=['#ff9999', '#66b3ff'])
plt.title('Learner Status Distribution')
plt.ylabel('')
plt.show()
```

```python
completed_count = df[df['Status'] == "Completed"].shape[0]
total_count = df.shape[0]
completed_percentage = (completed_count / total_count) * 100
print(f'Percentage of learners who have completed the course: {completed_percentage:.2f}%')
```
